```python
import pytrends
import country_abbrev
from pytrends.request import TrendReq
import pandas as pd
```

```python
import pandas as pd

df = pd.read_csv('../../../data.csv').drop('Unnamed: 0', axis=1)
```

```python
ax = df[(df.Country_o == 'Venezuela') & (df.year >= 2015) & (df.year <= 2020)].groupby('year').sum('newarrival').plot(y='newarrival', title='Refugee Departures from Venezuela')
ax.get_legend().remove()
ax.plot()
```

```python
ax = df[(df.Country_d == 'Colombia') & (df.year >= 2015) & (df.year <= 2020)].groupby('year').sum('newarrival').plot(y='newarrival', title='Refugee Arrivals in Colombia')
ax.get_legend().remove()
ax.plot()
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
pytrends.build_payload(kw_list=['refugiado','asilo'], geo=country_abbrev.country_to_abbrev['Colombia'], timeframe= '2015-01-01 2021-01-01')
colombia = pytrends.interest_over_time().reset_index().groupby(pd.Grouper(freq='2MS', key='date')).mean().drop('isPartial',axis=1)
colombia.plot(title='Google Trends in Colombia')
```

```python
country = 'Venezuela (Bolivarian Republic of)'
pytrends.build_payload(kw_list=['refugiado','asilo'], geo=country_abbrev.country_to_abbrev[country], timeframe= '2015-01-01 2021-01-01')
brazil_df = pytrends.interest_over_time().reset_index().groupby(pd.Grouper(freq='2MS', key='date')).mean().drop('isPartial',axis=1)
brazil_df.plot(title='Google Trends in Venezuela')
```
