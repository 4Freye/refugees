```python
import pytrends
import country_abbrev
from pytrends.request import TrendReq
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
pytrends.build_payload(kw_list=['migration', 'refugee'], geo=country_abbrev.country_to_abbrev['Mexico'], timeframe= 'all')
pytrends.interest_over_time()
```
